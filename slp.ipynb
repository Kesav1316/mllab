Aim:

PERFORM SINGLE LAYER PERCEPTRON ON LINEARLY SEPERABLE AND NON LINEARLY SEPERABLE DATA.

Algorithm:

The Single-Layer Perceptron (SLP) is a fundamental neural network model used for binary classification of linearly separable data. It consists of input neurons connected to an output neuron via weighted connections. The perceptron processes an input vector X=(x1,x2,...,xn)X=(x1​,x2​,...,xn​) by computing a weighted sum with corresponding weights W=(w1,w2,...,wn)W=(w1​,w2​,...,wn​), adding a bias bb, and passing the result through an activation function (typically a step function). Mathematically, the net input is computed as z=W⋅X+bz=W⋅X+b, and the output is determined as y^=f(z)y^​=f(z), where f(z)f(z) outputs 1 if zz is positive and 0 otherwise. The perceptron learns by adjusting the weights using the Perceptron Learning Rule, which updates weights when a misclassification occurs. If the predicted output y^y^​ differs from the actual label yy, the weights are updated using the formula wi=wi+η(y−y^)xiwi​=wi​+η(y−y^​)xi​ and the bias is updated as b=b+η(y−y^)b=b+η(y−y^​), where ηη is the learning rate. This process repeats over multiple iterations until the perceptron correctly classifies all training samples or reaches a stopping criterion. However, the single-layer perceptron can only learn linearly separable functions and fails on problems like XOR, which require non-linear decision boundaries. To address this, Multi-Layer Perceptrons (MLPs) with hidden layers and non-linear activation functions are used.

Code:



```python
#Linearly seperable data
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from sklearn import datasets
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset
iris = datasets.load_iris()
X = iris.data[:, [0, 2]]  # Selecting two features: sepal length & petal length
y = iris.target

# Step 2: Select two linearly separable classes (Setosa vs. Versicolor)
X = X[y < 2]  # Keep only class 0 (Setosa) and class 1 (Versicolor)
y = y[y < 2]  # Labels remain 0 and 1

# Step 3: Standardize the dataset (important for Perceptron training)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Step 4: Train a Single-Layer Perceptron
perceptron = Perceptron(max_iter=1000, eta0=0.01, random_state=42)
perceptron.fit(X, y)

# Step 5: Plot data points and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k', label="Data Points")

# Compute Decision Boundary: w1*x1 + w2*x2 + b = 0
x_values = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_values = -(perceptron.coef_[0][0] * x_values + perceptron.intercept_[0]) / perceptron.coef_[0][1]
plt.plot(x_values, y_values, 'k--', label='Decision Boundary')

plt.xlabel("Standardized Sepal Length")
plt.ylabel("Standardized Petal Length")
plt.title("Single-Layer Perceptron on Linearly Separable Iris Data")
plt.legend()
plt.show()
```

```python
#Non linearly seperable data
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Load the full Iris dataset
iris = datasets.load_iris()
X = iris.data[:, [0, 2]]  # Using sepal length & petal length (2 features)
y = iris.target  # Three classes: 0 (Setosa), 1 (Versicolor), 2 (Virginica)

# Step 2: Standardize the dataset
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Step 3: Train a Single-Layer Perceptron (SLP)
perceptron = Perceptron(max_iter=1000, eta0=0.01, random_state=42)
perceptron.fit(X, y)

# Step 4: Plot Data Points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k', label="Data Points")

# Step 5: Plot Decision Boundary
x_values = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_values = -(perceptron.coef_[0][0] * x_values + perceptron.intercept_[0]) / perceptron.coef_[0][1]
plt.plot(x_values, y_values, 'k--', label='Decision Boundary')

plt.xlabel("Standardized Sepal Length")
plt.ylabel("Standardized Petal Length")
plt.title("Single-Layer Perceptron on Non-Linearly Separable Iris Data")
plt.legend()
plt.show()
```
