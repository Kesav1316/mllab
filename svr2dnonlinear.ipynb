Aim:

Perform SVR on 2D non linearly seperable data

Algorithm:

Support Vector Regression (SVR) is a regression technique derived from Support Vector Machines (SVM), designed to predict continuous values. The goal of SVR is to find a function that best fits the data within a specified error margin (epsilon). It works by mapping input data into a higher-dimensional space using kernel functions (such as RBF or polynomial kernels) and then finding a hyperplane that minimizes the error while keeping the model as simple as possible. SVR introduces an epsilon margin, within which errors are tolerated, and only points outside this margin are considered as support vectors. The model is controlled by hyperparameters like C (which balances the trade-off between margin size and error), epsilon (which controls the tolerance for errors), and the kernel (which transforms the data). SVR is effective for both linear and non-linear data but can be computationally expensive and sensitive to the choice of hyperparameters. It is especially useful when dealing with high-dimensional or noisy data, offering robust performance with proper tuning.

Code:



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X, y = make_moons(n_samples=200, noise=0.1, random_state=42)

scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).ravel()
```

```python
svr = SVR(kernel='rbf', C=100, epsilon=0.01, gamma='auto')
svr.fit(X_scaled, y_scaled)

y_pred_scaled = svr.predict(X_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
```

Output:

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"RÂ² Score: {r2:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], y, color='blue', label="Actual Data", alpha=0.6)
plt.scatter(X[:, 0], y_pred, color='red', marker='x', label="SVR Predictions", alpha=0.7)
plt.title("Improved SVR on Non-Linear 2D Data (make_moons)")
plt.xlabel("Feature 1")
plt.ylabel("Target (y)")
plt.legend()
plt.show()
```

Result:

SVR has been applied on 2D non linearly seperable data.
